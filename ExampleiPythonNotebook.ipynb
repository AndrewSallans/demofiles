```python
import numpy as np
from text.classifiers import NaiveBayesClassifier
import pandas as pd
from text.blob import TextBlob as tb
import random
with open("~/data/couch.txt", 'r') as fp:
      couch = eval(fp.read())

text = []
val = []
for i in couch:
    text.append(str(i[0]))
    if i[1]>279:
        val.append("high")
    else:
        val.append("low")

                
# 694 items
x = [i for i in range(0,len(val))]

random.shuffle(x)

train = []
for i in range(400):
    i = x.pop()
    train.append((text[i], val[i]))    
  
    
    
test = []
for i in range(200):
    i = x.pop()
    test.append((text[i], val[i]))    
    

cl = NaiveBayesClassifier(train)
cl.accuracy(test)  

print("Accuracy: {0}".format(cl.accuracy(test)))

cl.show_informative_features(25)
    
```
